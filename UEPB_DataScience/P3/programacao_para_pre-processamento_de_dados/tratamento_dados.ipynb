## Importando arquivo

```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"Banco de dados filmes(TB_Principal).csv", sep=";")
df.head()
```

```python
#limpando colunas vazias

df = df.drop(columns=["Unnamed: 13", "Unnamed: 14"])
```

```python
#Convertendo colunas numéricas

# Função para limpar valores monetários
def limpar_valor(valor):
    if pd.isna(valor):
        return np.nan
    return float(valor.replace("$", "").replace(".", "").replace(",", "."))

df["Orçamento"] = df["Orçamento"].apply(limpar_valor)
df["Receita"] = df["Receita"].apply(limpar_valor)
```

## Importando dados e bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura do arquivo CSV
arquivo = "megalhaes.csv"
df = pd.read_csv(arquivo, sep=",", encoding="utf-8")

# Exibir informações básicas
print("Dimensão do dataset:", df.shape)
print("\nColunas disponíveis:")
print(df.columns.tolist())

# Visualizar as primeiras linhas do conjunto de dados
df.head()
```

## Gráfico 1 - Peso x Exercício

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='exercicio', y='peso', hue='sexo', style='sexo', s=80)
plt.title('Relação entre Peso e Horas de Exercício por Semana')
plt.xlabel('Horas de Exercício por Semana')
plt.ylabel('Peso (kg)')
plt.legend(title='Sexo')
plt.show()
```

Diferença de Peso por Gênero: O peso médio dos indivíduos do sexo masculino (M) é notavelmente superior ao do sexo feminino (F) na amostra.

Ausência de Correlação: Não se observa uma correlação linear clara entre as horas de exercício e o peso para a amostra como um todo ou para cada gênero isoladamente. O peso varia amplamente, independentemente da quantidade de exercício semanal, no intervalo de 0 a 8 horas.

Valores Extremos: Os pesos mais elevados (>90 kg) e os mais baixos (≈45 kg) aparecem em ambas as extremidades do espectro de exercício (0 e 10 horas), sugerindo que outros fatores além do exercício semanal influenciam o peso.

## Gráfico 2 - Opinião sobre cinema - tolerância a cigarro

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='opcinema', hue='tolerancia')
plt.title('Opinião sobre Cinema por Nível de Tolerância ao Cigarro')
plt.xlabel('Opinião sobre o Cinema (B = Boa, M = Muito boa)')
plt.ylabel('Quantidade de alunos')
plt.legend(title='Tolerância (I, P, M)')
plt.show()
```

Opinião Prevalente: A opinião "Boa" (B) sobre cinema é a mais frequente, totalizando quase o dobro da opinião "Muito Boa" (M).

Distribuição da Tolerância: As categorias Pouco Tolerante (P) e Médio Tolerante (M) são as mais representadas em ambas as opiniões sobre cinema ("Boa" e "Muito Boa"). Isso pode refletir a distribuição da tolerância na amostra total, onde os intolerantes (I) são o menor grupo.

Pouco Tolerante (P): Esta é a categoria mais frequente entre aqueles que dão a opinião "Muito Boa" (M), sugerindo que o grupo P tem uma ligeira tendência a classificar o cinema de forma mais positiva.

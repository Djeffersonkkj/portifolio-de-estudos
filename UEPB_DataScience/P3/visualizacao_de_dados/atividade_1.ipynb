**1**. Gere uma serie de 0 a 9 e plote uma linha. Ajuste os ticks do eixo x para 0, 2, 4,
6, 8 e limite o eixo x para [-1, 10] e insira o t ́ıtulo. Isso refor ̧ca que xticks/xlim
controlam marca ̧c ̃oes e limites.

2. Crie plt.subplots(2,2, sharex=True, sharey=True) e desenhe um histograma
em cada axes[i, j]. Em seguida, elimine o espa ̧camento com fig.subplots adjust(wspace=0,
hspace=0).

```python
# Dados aleatórios para os histogramas
dados = np.random.randn(1000)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i, j].hist(dados, bins=20, color='skyblue', edgecolor='black')
        
fig.subplots_adjust(wspace=0, hspace=0)
```

3. Crie uma pd.Series de 16 valores e plote barras verticais e depois horizontais.
Lembre que plot.bar() e plot.barh() fazem as duas variantes e o  ́ındice  ́e usado
nas marca ̧c ̃oes.

```python
valores = pd.Series(np.random.randint(1, 20, 16))

plt.figure(figsize=(8, 4))
valores.plot.bar(color='skyblue', edgecolor='black')
plt.title("Barras Verticais")
plt.show()

plt.figure(figsize=(8, 4))
valores.plot.barh(color='orange', edgecolor='black')
plt.title("Barras Horizontais")
plt.show()
```

5. Crie fig, axes = plt.subplots(1,2) e desenhe a mesma s ́erie em cada painel,
mas com estilos distintos, usando o parˆametro ax=.

```python
x = np.arange(10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y, color='blue', marker='o', linestyle='-')
axes[0].set_title("Estilo 1: linha contínua com pontos")

axes[1].plot(x, y, color='red', marker='s', linestyle='--')
axes[1].set_title("Estilo 2: linha tracejada com quadrados")

plt.tight_layout()
plt.show()
```

6. Plote um histograma e, por cima, uma linha; use alpha para deixar a sobreposi ̧c ̃ao
vis ́ıvel.

```python
np.random.seed(42)
dados6 = np.random.normal(loc=50, scale=10, size=200)
contagens, bordas = np.histogram(dados6, bins=15, density=True)
centros = (bordas[:-1] + bordas[1:]) / 2

plt.hist(dados6, bins=15, density=True, alpha=0.5, edgecolor='black')
plt.plot(centros, contagens, linewidth=2, alpha=0.9)

plt.xlabel("Valor")
plt.ylabel("Densidade")
plt.title("Histograma com linha sobreposta")
```

7. Fa ̧ca duas linhas: uma com cor nomeada ("steelblue"), outra com um hex ("#CECECE").
Explique o que \#CECECE" representa em RGB.

```python
x_7 = np.linspace(0, 10, 100)
y1_7 = np.sin(x_7)
y2_7 = np.cos(x_7)

plt.plot(x_7, y1_7, color="steelblue", label="Seno (steelblue)")
plt.plot(x_7, y2_7, color="#CECECE", label="Cosseno (#CECECE)")

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Duas linhas com cores diferentes")
plt.legend()
```

8. Crie dois subplots com sharex=True. Depois de plotar seno e cosseno, chame
axes[0].set xlim(0, 2*np.pi) e observe que o limite  ́e compartilhado. (Com
sharex, ajustar xlim impacta todas as subplots.)

```python
x = np.linspace(0, 4*np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axes = plt.subplots(2, 1, sharex=True, figsize=(6, 5))

axes[0].plot(x, y_sin, color="steelblue", label="Seno")
axes[0].legend()
axes[0].set_title("Função Seno")

axes[1].plot(x, y_cos, color="orange", label="Cosseno")
axes[1].legend()
axes[1].set_title("Função Cosseno")

axes[0].set_xlim(0, 2*np.pi)

plt.tight_layout()
plt.show()
```

9. Crie uma figura 2×2 via plt.figure e adicione trˆes subplots com fig.add subplot(2,2,i),
usando uma barra, um histograma e um scatter.

```python
np.random.seed(42)
valores_barra = [5, 7, 3, 4]
categorias = ['A', 'B', 'C', 'D']
dados_hist = np.random.randn(200)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

fig = plt.figure(figsize=(8, 6))

# Subplot 1: Barra
ax1 = fig.add_subplot(2, 2, 1)
ax1.bar(categorias, valores_barra, color='steelblue')
ax1.set_title("Gráfico de Barras")

# Subplot 2: Histograma
ax2 = fig.add_subplot(2, 2, 2)
ax2.hist(dados_hist, bins=15, alpha=0.7, edgecolor='black')
ax2.set_title("Histograma")

# Subplot 3: Scatter
ax3 = fig.add_subplot(2, 2, 3)
ax3.scatter(x_scatter, y_scatter, color='orange', edgecolor='black')
ax3.set_title("Gráfico de Dispersão (Scatter)")

plt.tight_layout()
```

10. Gere 100 categorias aleat ́orias entre ["A","B","C","D"] e plote as contagens com
s.value counts().plot.bar(). (O  ́ındice vira a marca ̧c ̃ao do eixo nas barras.)

```python
np.random.seed(42)
s = pd.Series(np.random.choice(["A", "B", "C", "D"], size=100))

# Contagem e gráfico de barras
s.value_counts().plot.bar(color='steelblue', edgecolor='black')

plt.xlabel("Categorias")
plt.ylabel("Frequência")
plt.title("Contagem de Categorias")
```

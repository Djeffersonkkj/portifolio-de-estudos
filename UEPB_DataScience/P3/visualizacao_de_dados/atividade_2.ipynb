```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
#salvando arquivo .cvs
data = pd.read_csv(
    "/Users/djeff/OneDrive/Documents/MeusProjetos/portifolio-de-estudos/UEPB_DataScience/P3/visualizacao_de_dados/dados_atvdd_2.csv", sep=";" #tive que colocar isso pois o csv não veio em UTF-8
)

cols = ['Claridade', 'Aroma', 'Corpo', 'Sabor', 'Afinação', 'Qualidade']

# Substituir vírgula por ponto e converter para float
for col in cols:
    data[col] = pd.to_numeric(data[col].astype(str).str.replace(',', '.'), errors='coerce')

data.head()
```

#### **(c) Obtenha a matrix de diagrama de dispersão e indique em quais diagramas há o indicativo de possíveis relações lineares entre as features (características).**

```python
scatter_matrix(data, figsize=(12, 12), diagonal='hist', alpha=0.7, marker='o')
```

O gráfico indica que Claridade praticamente não se conecta com as outras características. Já Aroma, Corpo e Sabor tendem a crescer juntos e influenciam um pouco a Qualidade, que é a que mais sofre essa influência. A relação mais clara é entre Afinação e Qualidade, enquanto os outros padrões são mais sutis.

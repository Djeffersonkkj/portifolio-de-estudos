## Importando dados e bibliotecas

```python
# Importando as bibliotecas principais
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Importando base de dados

arquivo = "PB Pescadores(in).csv"
df = pd.read_csv(arquivo, encoding='utf-8', sep=',')
df.head()
```

```python
#Informações dos dados

df.info()
df.isnull().sum()
df.describe(include='object')
```

## Escolhendo Variáveis Qualitativas

Com base na análise exploratória inicial, observou-se que a variável **“Categoria”** apresenta quase todos os registros como “Artesanal”,
não oferecendo variação significativa para fins analíticos.  

Assim, optou-se por selecionar as seguintes variáveis qualitativas:

1. **Sexo** – indica o gênero do pescador ou pescadora profissional;  
2. **Nível Escolaridade** – reflete o grau de instrução;  
3. **Forma de Atuação** – distingue se o trabalho é embarcado ou desembarcado.

Essas variáveis foram escolhidas por oferecerem diversidade de categorias e
permitirem identificar padrões sociais e profissionais na comunidade pesqueira da Paraíba.

## Gerando gráficos

As análises realizadas foram:

1. **Sexo × Nível Escolaridade:** mostra diferenças no grau de instrução entre homens e mulheres.  
2. **Sexo × Forma de Atuação:** mostra se há predominância de embarque/desembarque por gênero.  
3. **Forma de Atuação × Nível Escolaridade:** analisa se a escolaridade influencia o tipo de atuação na pesca.

```python
# ---------- 1. Sexo x Nível Escolaridade ----------
tabela1 = pd.crosstab(df['Nivel Escolaridade'], df['Sexo'])
tabela1.plot(kind='bar', figsize=(10,5))
plt.title('Distribuição do Nível de Escolaridade por Sexo')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Pescadores')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

```python
# ---------- 2. Sexo x Forma de Atuação ----------
tabela2 = pd.crosstab(df['Forma de Atuacao'], df['Sexo'])
tabela2.plot(kind='bar', figsize=(6,5), color=['#4C72B0','#55A868'])
plt.title('Distribuição da Forma de Atuação por Sexo')
plt.xlabel('Forma de Atuação')
plt.ylabel('Quantidade de Pescadores')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

```python
# ---------- 3. Forma de Atuação x Nível Escolaridade ----------
tabela3 = pd.crosstab(df['Nivel Escolaridade'], df['Forma de Atuacao'])
tabela3.plot(kind='bar', figsize=(10,5))
plt.title('Distribuição do Nível de Escolaridade por Forma de Atuação')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Pescadores')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Forma de Atuação')
plt.tight_layout()
plt.show()
```

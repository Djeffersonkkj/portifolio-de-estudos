```python
#Importando Bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Importando dados

data = {
    "Alturas": [1.78, 1.79, 1.67, 1.56, 1.63, 1.80, 2.00, 1.50, 1.75, 1.54, 
                1.79, 1.72, 1.63, 1.65, 1.58, 1.39, 1.75, 1.82, 1.93, 1.98],
    "Preferencia_musical": ["F", "F", "F", "R", "F", "R", "R", "R", "S", "S",
                            "R", "S", "F", "R", "F", "S", "R", "F", "S", "R"],
    "Frequencia_transporte": ["as_vezes", "nunca", "nunca", "nunca", "as_vezes",
                              "frequentemente", "frequentemente", "frequentemente", "nunca", "nunca",
                              "as_vezes", "as_vezes", "as_vezes", "as_vezes", "as_vezes", "as_vezes",
                              "nunca", "nunca", "nunca", "frequentemente"]
}

df = pd.DataFrame(data)
```

### **QUESTÃO 1**

   O tipo de gráfico escolhido foi o **Histograma**, pois a variável *altura* é contínua. Portanto, esse formato é o mais adequado para mostrar a distribuição de frequência.

```python
# Histograma das Alturas
plt.figure(figsize=(8,5))
sns.histplot(df["Alturas"], bins=8, kde=True, color="skyblue", edgecolor="black")
plt.title("Distribuição das Alturas dos Participantes")
plt.xlabel("Altura (m)")
plt.ylabel("Frequência")
plt.show()
```

A maioria dos participantes tem altura concentrada entre 1,65m e 1,85m, sugerindo um padrão centralizado em torno de valores médios. Existem poucos valores mais extremos (como 1,39m e 2,00m), que podem ser considerados outliers dentro da amostra.

### **QUESTÃO 2**

A variável **Preferencia_musical** é categórica nominal, logo, o gráfico de barras é ideal para comparar a quantidade de participantes em cada categoria.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Preferencia_musical", data=df, palette="pastel")
plt.title("Preferência Musical dos Participantes")
plt.xlabel("Preferência Musical")
plt.ylabel("Contagem")
```

A categoria R (Rock) aparece como a mais comum (8 ocorrências), seguida de F (Forro) (7 ocorrências) e S (Sertanejo) (5 ocorrências). Isso mostra que existe uma predominância de pessoas que preferem Rock nesta amostra ou uma predominância nas pessoas que não gostam de rock.

### **QUESTÃO 3**

A variável **Frequencia_transporte** também é categórica nominal, logo, a contagem em barras é a melhor forma de representar a frequência de cada categoria.

```python
# Gráfico de Barras - Frequência de Transporte
plt.figure(figsize=(7,5))
sns.countplot(
    x="Frequencia_transporte", 
    data=df, 
    palette="muted", 
    order=df["Frequencia_transporte"].value_counts().index
)
plt.title("Frequência de Uso do Transporte Público")
plt.xlabel("Frequência")
plt.ylabel("Contagem")
plt.show()
```

As categorias "as_vezes" e "nunca" aparecem com a mesma frequência (8 cada), enquanto "frequentemente" foi menos comum (4). Isso sugere que a maioria dos participantes não utiliza o transporte público de forma frequente.

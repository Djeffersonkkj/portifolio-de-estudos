### 4. Dez alunos foram submetidos a um teste de Estatística e de Matemática, obtendo as seguintes notas:

#### e) Represente graficamente a reta

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Dados
X = np.array([6, 5, 9, 10, 3, 4, 8, 7, 6, 2])
Y = np.array([7, 6, 10, 9, 2, 3, 9, 5, 6, 3])

# (a) Covariância
cov_matrix = np.cov(X, Y, ddof=1)  # ddof=1 para amostra
cov_xy = cov_matrix[0, 1]

# (b) Variâncias
var_x = np.var(X, ddof=1)
var_y = np.var(Y, ddof=1)

# (c) Coeficiente de correlação
corr_xy = np.corrcoef(X, Y)[0, 1]

# (d) Coeficientes do modelo de regressão linear simples
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

# (e) Representação gráfica
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color="blue", label="Dados")
plt.plot(X, intercept + slope * X, color="red", label=f"y = {intercept:.2f} + {slope:.2f}x")
plt.xlabel("Matemática (X)")
plt.ylabel("Estatística (Y)")
plt.title("Regressão Linear Simples")
plt.legend()
plt.grid(True)
plt.show()

# Resultados
print(f"(a) Covariância: {cov_xy:.2f}")
print(f"(b) Variância de X: {var_x:.2f}, Variância de Y: {var_y:.2f}")
print(f"(c) Coeficiente de correlação: {corr_xy:.2f}")
print(f"(d) Modelo de regressão linear: Y = {intercept:.2f} + {slope:.2f}X")
```

### 6. É esperado que a massa muscular de uma pessoa diminua com a idade. Para estudar essa relação, uma nutricionista selecionou 18 mulheres, com idade entre 40 e 79 anos, e observou em cada uma delas a idade (X) e a massa muscular (Y).

```python
# Importar bibliotecas e definir dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Dados
dados = {
    "Massa_muscular_Y": [82.0, 91.0, 100.0, 68.0, 87.0, 73.0, 78.0, 80.0, 65.0, 84.0, 116.0, 76.0, 97.0, 100.0, 105.0, 77.0, 73.0, 78.0],
    "Idade_X": [71.0, 64.0, 43.0, 67.0, 56.0, 73.0, 68.0, 56.0, 76.0, 65.0, 45.0, 58.0, 45.0, 53.0, 49.0, 78.0, 73.0, 68.0]
}

df = pd.DataFrame(dados)
```

#### a) Construa o diagrama de dispersão e interprete-o.

```python
# Diagrama de dispersão
plt.figure(figsize=(8,6))
sns.scatterplot(x="Idade_X", y="Massa_muscular_Y", data=df, color="blue", s=80)
plt.title("Diagrama de Dispersão: Massa Muscular vs Idade")
plt.xlabel("Idade (anos)")
plt.ylabel("Massa Muscular")
plt.grid(True)
plt.show()
```

Interpretação:
O gráfico de dispersão mostra uma tendência de massa muscular diminuir com o aumento da idade. Há uma relação negativa aparente entre Idade e Massa Muscular.

#### b) Determinar a covariância de X e Y

```python
# Covariância
covariancia = np.cov(df["Idade_X"], df["Massa_muscular_Y"], ddof=1)[0,1]
print(f"Covariância entre X e Y: {covariancia:.2f}")
```

#### c) Calcule o coeficiente de correlação linear entre X e Y.

```python
# Coeficiente de correlação linear
correlacao, _ = pearsonr(df["Idade_X"], df["Massa_muscular_Y"])
print(f"Coeficiente de correlação linear (r): {correlacao:.4f}")
```

#### d) Determine os coeficientes do modelo de regressão linear simples

```python
# Modelo de regressão linear simples
from sklearn.linear_model import LinearRegression

X = df["Idade_X"].values.reshape(-1,1)
y = df["Massa_muscular_Y"].values

modelo = LinearRegression()
modelo.fit(X, y)

b0 = modelo.intercept_
b1 = modelo.coef_[0]

print(f"Modelo de regressão linear: Y = {b0:.2f} + {b1:.2f}X")
```

#### e) Represente graficamente a reta

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Idade_X", y="Massa_muscular_Y", data=df, color="blue", s=80, label="Dados")
plt.plot(df["Idade_X"], modelo.predict(X), color="red", linewidth=2, label="Reta de Regressão")
plt.title("Reta de Regressão Linear: Massa Muscular vs Idade")
plt.xlabel("Idade (anos)")
plt.ylabel("Massa Muscular")
plt.legend()
plt.grid(True)
plt.show()
```

### 10. A tabela representa uma amostra com os pesos de 10 pais e seus filhos mais velhos.

```python
dados = {
    "Pais": [60, 65, 70, 68, 63, 69, 71, 64, 66, 64],
    "Filhos": [63, 64, 71, 69, 63, 68, 73, 63, 64, 62]
}

df = pd.DataFrame(dados)
```

#### a) Calcular o coeficiente de correlação entre os pesos dos pais e dos filhos utilizando as seguintes transformações das variáveis Z = (X - 66) e W = (Y - 67)

```python

df["Z"] = df["Pais"] - 66
df["W"] = df["Filhos"] - 67

df.head()

```

```python
correlacao = df["Z"].corr(df["W"])
print(f"Coeficiente de correlação entre Z e W: {correlacao:.4f}")
```

#### b) Calcule os seguintes desvios-padrão Sx, Sz, Sy, Sw


```python
std_X = df["Pais"].std(ddof=1)
std_Y = df["Filhos"].std(ddof=1)
std_Z = df["Z"].std(ddof=1)
std_W = df["W"].std(ddof=1)

print(f"Desvio-padrão de X (Pais): {std_X:.4f}")
print(f"Desvio-padrão de Y (Filhos): {std_Y:.4f}")
print(f"Desvio-padrão de Z: {std_Z:.4f}")
print(f"Desvio-padrão de W: {std_W:.4f}")
```

#### c) Determinar o valor de b, coeficiente angular da reta :

```python
cov_ZW = np.cov(df["Z"], df["W"], ddof=1)[0,1]
b = cov_ZW / (std_Z**2)

print(f"Coeficiente angular (b): {b:.4f}")
```

```python
media_X = df["Pais"].mean()
media_Y = df["Filhos"].mean()

a = media_Y - b * media_X
print(f"Coeficiente linear (a): {a:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(df["Pais"], df["Filhos"], color="blue", label="Dados")
plt.plot(df["Pais"], a + b*df["Pais"], color="red", label="Reta de regressão")
plt.xlabel("Peso dos pais (X)")
plt.ylabel("Peso dos filhos (Y)")
plt.title("Reta de regressão: Y sobre X")
plt.legend()
plt.grid(True)
plt.show()
```
